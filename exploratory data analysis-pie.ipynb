```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("shopping_trends.csv")

# Exploratory Data Analysis (EDA) - Pie Chart Most Context for Nonconditional Data
target_col = ['Gender','Item Purchased','Category','Location','Color','Season','Payment Method','Shipping Type','Discount Applied','Preferred Payment Method','Frequency of Purchases']

for dt in target_col:
    # Dataset
    top_context = df[dt].value_counts().head(7)
    
    # Pie Chart
    plt.figure(figsize=(8,8))
    top_context.plot.pie(autopct='%1.1f%%', startangle=90)
    plt.title(f"Most {dt}")
    plt.ylabel("")
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("shopping_trends.csv")
df

# Others Asset
regions = {
    "Cash": ["Cash"],
    "E-Payment": ["Credit Card", "Bank Transfer", "PayPal", "Venmo", "Debit Card"]
}

def assign_region(state):
    for region, states in regions.items():
        if state in states:
            return region
    return "Unknown"

# Implement Payment Type Based On Payment Method
df["Payment Type"] = df["Payment Method"].apply(assign_region)

# Data Prepare - List Unregistered Payment Type
unknown_region = df[df["Payment Type"] == "Unknown"]
print(unknown_region[["Payment Type","Payment Method"]])
```

```python
# Exploratory Data Analysis (EDA) - Pie Chart Comparison Payment Type
# Dataset
top_context = df["Payment Type"].value_counts().head(7)

# Pie Chart
plt.figure(figsize=(8,8))
top_context.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title(f"Comparison Payment Type")
plt.ylabel("")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("shopping_trends.csv")
df

# Others Asset
shipping_type = {
    "No": ["Store Pickup"],
    "Yes": ["Express","Free Shipping","Next Day Air","Standard","2-Day Shipping"]
}

def assign_shipping_type(types):
    for dt_type, types_list in shipping_type.items():
        if types in types_list:
            return dt_type
    return "Unknown"

# Implement Is Shipping Based On Shipping Type
df["Is Shipping"] = df["Shipping Type"].apply(assign_shipping_type)

# Data Prepare - List Unregistered Type
unknown_type = df[df["Shipping Type"] == "Unknown"]
print(unknown_type[["Is Shipping","Shipping Type"]])
```

```python
# Exploratory Data Analysis (EDA) - Pie Chart Comparison Is Shipping?
# Dataset
top_context = df["Is Shipping"].value_counts()

# # Pie Chart
plt.figure(figsize=(8,8))
top_context.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title(f"Comparison Is Shipping?")
plt.ylabel("")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("shopping_trends.csv")
df

# Others Asset
color_type = {
    "Light": ["White", "Silver", "Pink", "Lavender", "Peach", "Cyan", "Beige", "Orange", "Yellow", "Magenta", "Gold"],
    "Dark": ["Gray", "Maroon", "Turquoise", "Charcoal", "Purple", "Olive", "Teal", "Black", "Green", "Red", "Brown", "Indigo", "Blue","Violet"]
}

def assign_color_type(types):
    for dt_type, types_list in color_type.items():
        if types in types_list:
            return dt_type
    return "Unknown"

# Implement Is Color Type On Color
df["Color Type"] = df["Color"].apply(assign_color_type)

# Data Prepare - List Unregistered Type
unknown_type = df[df["Color Type"] == "Unknown"]
print(unknown_type[["Color","Color Type"]])
```
